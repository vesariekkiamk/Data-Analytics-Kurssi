# Luon Dictionaryn eri info tyypeillä.

```python
Dictionary = {"Discrete":[1, 2, 3, 4], "Continuous":[12.5, 16.0, 20.2, 30.4], "Nominal":["man", "woman", "child", "animal"], "Ordinal":["yes", "no", "maybe", "no opinion"]}
```

## Testaan .get toimintoa.

```python
print(Dictionary.get("Discrete"))
```

### Lopuksi kokeilen omaa koodia joka hakee tietyn datan listasta ja printtaa tekstin sen perusteella.

```python
print(Dictionary.get("Nominal"))

if "plane" in Dictionary["Nominal"]:
    print("found")
else:
    print("not found")

if "man" in Dictionary["Nominal"]:
    print("found")
else:
    print("not found")
```
