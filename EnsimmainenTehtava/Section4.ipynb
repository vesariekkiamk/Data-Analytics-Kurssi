#### Lisätään Section 3 kohdasta Dictioary ja lisätään "Nominaliin" yksi toistuva data merkintä.

```python
Dictionary = {"Discrete":[1, 2, 3, 4], "Continuous":[12.5, 16.0, 20.2, 30.4], "Nominal":["man", "woman", "child", "animal", "animal"], "Ordinal":["yes", "no", "maybe", "no opinion"]}
```

#### Testataan statistic.modea joka hakee eniten toistuvan datan.

```python
statistics.mode(Dictionary["Nominal"])
```

```python
statistics.mode(Dictionary["Ordinal"])
```

#### "yes" merkintä tulee sen takia ylemmässä koodissa, koska ei ole eniten toistuvaa dataa niin koodi ottaa ensimmäinen arvon jonka se saa.

## Seuraavaksi kokeillaan statistic.mean komentoa joka hakee keskiarvon.

```python
statistics.mean(Dictionary["Continuous"])
```

```python
statistics.mean(Dictionary["Discrete"])
```

##### Tässä kohtaa testaillaan koodilla huvikseen taas eri print vaihtoehtoja tuloksien mukaan.

```python
arvo = statistics.mean(Dictionary["Discrete"])

if arvo <= (2.4):
    print("liian pieni arvo")
else:
    print("iso arvo")
```
