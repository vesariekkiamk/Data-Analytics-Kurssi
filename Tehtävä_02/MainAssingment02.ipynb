## **1.5**

### Low nopan pienin luku, high nopan isoin luku ("20") ja size nopan heittojen määrä.

```python
d20_roolipelinoppa = np.random.randint(low=1, high=21, size=100)
print(d20_roolipelinoppa)
```

### Bincount() laskee lukujen 1-20 määrät sadan heiton aikana.

```python
määrät = np.bincount(d20_roolipelinoppa)
print(määrät)
```

### Seuraava kohta kertoo printillä hieman selvemmin lukujen määrät.

```python
for arvo in range (1, 21):
    print(f"Luku {arvo} toistuu {määrät[arvo]} kertaa")
```

## **2.1**

```python
import pandas as pd
```

### Encodataan jotta pystytään lukemaan Euroopan kielistä dataa.

```python
df = pd.read_csv('koeti_riista_tau_001_fi.csv', encoding='8859')
```

### Löydetään jokaisen kuukauden onnettomuuksien määrät ja sitten yksi kuukausi jolloin niitä on eniten.

```python
onnettomuuksienMäärä = df.groupby('kuukausi')['id'].sum()
onnettomuuksienMäärä.plot(kind="bar")
```

### Vuoden alkupuolella on vähiten onnettomuuksia ja niiden määrä lähtee kasvuun vuoden edetessä. Eniten onnettomuuksia on Marraskuussa.

## **2.2**

### Käytetään crosstab komentoa jotta saadaan luotua taulukko jostaa näkee kaikkien kuntien eri eläinten aiheuttamat tilanteet. Margins=True lisää "All" kohdan joka kertoo yhteensä kaikkien onnetomuuksien määrän kunnassa.

```python
taulukko = pd.crosstab(df['maakuntaNimi'], df['riistalajiNimi'], margins=True)
print(taulukko)
```

------------------------

## **2.3**

### Convertoidaan Pandasin aikaformaattiin ja käännetään Helsingin aikaan. Tämän selvittämisessä meni aikaa.

```python
milloin = df["tapahtumaAika"] = pd.to_datetime(df["tapahtumaAika"], utc=True).dt.tz_convert('Europe/Helsinki')
```

### Luodaan viikonpäivät ja tunnit.

```python
df["tuntiPäivästä"] = df["tapahtumaAika"].dt.hour
```

```python
df["viikonpäivä"] = df["tapahtumaAika"].dt.day_name()
```

### En koskaan saanut selvitettyä miten saan koodin palauttamaan pelkän päivämäärän ja kellonajan. Koodi ei pitänyt "%d" merkinnästä vaan halusi minun käyttävän isoa d-kirjainta jonka takia printti on tuollainen.

```python
print(milloin.dt.strftime("%D.%m.%Y %H:%M"))
```

### Seuraava kohta kertoo meille onnettomuuksien määrän eri tuntien ajalle.

```python
onnettomuuksienMääräTunnissa = df["tuntiPäivästä"].value_counts().sort_index()
```

```python
print(onnettomuuksienMääräTunnissa)
```

### Otetaan esiin info tunnista jolloin tapahtuu eniten onnettomuuksia.

```python
print(onnettomuuksienMääräTunnissa.idxmax())
```
